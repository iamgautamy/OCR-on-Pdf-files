```python
import cv2
import pytesseract
import numpy as np
from pdf2image import convert_from_path
```

```python
pdf = r'D:\self_project_frameworks\OCR\testpdf.pdf'
pages = convert_from_path(pdf, 350)
i = 1
for page in pages:
    image_name = "D:\self_project_frameworks\OCR\gg\\" + str(i) + ".jpg"  
    page.save(image_name, "JPEG")
    i = i+1 
```

```python
import cv2
import pytesseract

img = cv2.imread(r'D:\self_project_frameworks\OCR\gg\1.jpg')

h, w, c = img.shape
for i in range(1,len(pages)):
    path = "D:\self_project_frameworks\OCR\gg\*.*"
    for file in glob.glob(path):
        image_read = cv2.imread(file)
        boxes = pytesseract.image_to_boxes(image_read) 
        for b in boxes.splitlines():
            b = b.split(' ')
            img = cv2.rectangle(image_read, (int(b[1]), h - int(b[2])), (int(b[3]), h - int(b[4])), (0, 255, 0), 2)
        image_names = "D:\self_project_frameworks\OCR\output_pdf\\"+"Page_" + str(i) + ".jpg"
        i+=1
        cv2.imwrite(image_names, img)     
    
```
